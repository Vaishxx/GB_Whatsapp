```python
import pandas as pd
import re
```

```python
f=open('WhatsApp Chat with SDSF, DAVV.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APap][Mm]\s-\s'
```

```python
msg=re.split(pattern,data)[1:]
msg
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_msg':msg,'msg_date':dates})
#convert mdg_date type
df['msg_date']=pd.to_datetime(df['msg_date'],format='%m/%d/%y, %I:%M %p - '
)
df.rename(columns={'msg_date':'date'},inplace=True)
# df['date_12hr']=df['date'].dt.strftime('%I:%M %p')
df.head()
```

```python
df.shape
```

```python
#separate users and messages
users=[]
messages=[]
for msg in df['user_msg']:
    entry=re.split(r'([\W\W]+?):\s',msg)
    if entry[1:]:#user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['users']=users
df['msg']=messages
df.drop(columns=['user_msg'],inplace=True)
df.head(20)
```

```python
df['year']=df['date'].dt.year
```
